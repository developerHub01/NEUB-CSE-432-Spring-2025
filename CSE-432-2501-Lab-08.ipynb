## Task 1
Manually calculating DFT of a discrete time sequence 

```python
import numpy as np
import matplotlib.pyplot as plt
xt=np.array([1,2,3,4])
N=len(xt)
Xw=np.zeros(N, dtype=complex)

for k in range(N):
    for n in range(N):
        Xw[k]=Xw[k]+xt[n]*np.exp(-2j*np.pi*k*n/N)

print(Xw)
```

## Task 2
Calculating DFT of a sinusoid using FFT function of numpy

```python
import numpy as np
import matplotlib.pyplot as plt

f=50 #Frequency of sinusoid f=50hz
Fs=1000 #Sampling frequency of 1000hz
T=1 #Signal duration of 1s

t=np.linspace(0,T,int(Fs*T), endpoint=False)
x=np.sin(2*np.pi*f*t)
print(len(x))

plt.figure()
plt.plot(t[0:256],x[0:256])
plt.title("Original sinusoid of 50Hz")
plt.xlabel("time t")
plt.ylabel("$sin(2*\pi*f*t)$")

Xw=np.fft.fft(x)
N=len(x)
frequencies=np.fft.fftfreq(N, d=1/Fs)

plt.figure()
plt.plot(frequencies[:N//2],np.abs(Xw[:N//2])/N)
plt.title("FFt of sinusoid of 50Hz")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```

## Task 3
Manually calculating DFT of a sinusoid

```python
import numpy as np
import matplotlib.pyplot as plt

f1=100
f2=200
f3=350
Fs=1000 #Sampling frequency of 1000hz
T=1 #Signal duration of 1s
L=64 #No of sampe to visualize

t=np.linspace(0,T,int(Fs*T), endpoint=False)
x1=np.sin(2*np.pi*f1*t)
x2=np.sin(2*np.pi*f2*t)
x3=np.sin(2*np.pi*f3*t)
x=x1+x2+x3

plt.figure()
plt.plot(t[0:L],x[0:L])
# plt.stem(np.arange(L),x[0:L])
plt.title("Original Signal")
plt.xlabel("time t")
plt.ylabel("x(t)")

N=len(x)
Xw=np.zeros(N,dtype=complex)

for k in range(N):
    for n in range(N):
        Xw[k]=Xw[k]+x[n]*np.exp(-2j*np.pi*k*n/N)
frequencies=np.linspace(0,Fs/2,N//2)
Y=np.abs(Xw[:N//2])/N

plt.figure()
plt.plot(frequencies,Y)
plt.title("Frequency Spectrum of Sinusoid using DFT")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
f1, f2, f3 = 100, 200, 350
Fs = 1000
T = 1
N = int(Fs * T)
L = 64

# Signal generation using loop
x = np.zeros(N)
for i in range(N):
    x[i] = np.sin(2 * np.pi * f1 * i/Fs) + np.sin(2 * np.pi * f2 * i /Fs) + np.sin(2 * np.pi * f3 * i /Fs)

# Plot original signal (first L samples)
plt.figure()
plt.plot(range(L), x[:L])
plt.title("Original Signal (Loop Generated)")
plt.xlabel("Sample number")
plt.ylabel("Amplitude")

# Manual DFT calculation
Xw = np.zeros(N, dtype=complex)
for k in range(N):
    for n in range(N):
        Xw[k] += x[n] * np.exp(-2j * np.pi * k * n / N)

# Frequency axis
frequencies = np.linspace(0, Fs/2, N//2)

# Magnitude spectrum (normalize by N)
Y = np.abs(Xw[:N//2]) / N

# Plot frequency spectrum
plt.figure()
plt.plot(frequencies, Y)
plt.title("Frequency Spectrum (Manual DFT)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```
