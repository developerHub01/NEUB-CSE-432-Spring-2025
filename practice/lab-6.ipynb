## Task 1
Downsampling

````python
import numpy as np
import matplotlib.pyplot as plt

x=[3,5,7,0,1,4,2] #Original Signal
n=np.arange(-2,5,1) 
plt.figure()
plt.stem(n,x)
plt.title('Original Signal x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')

dsFactor=2

y=x[0::dsFactor]
k=n[::dsFactor]

plt.figure()
plt.stem(k, y)
plt.title('Downsampled Signal x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.show()
````

This downsampling method just omits every other signal. Another method of downsampling can be done by grouping the signals into blocks and finding either mean or median.
```
Original Signal              3,5,7,0,1,4,2
Signals Divided into blocks  3,5   7,0  1,4  
Finding average of each block 4    3.5  2.5  
```
### Task 2 
Do downsampling by using mean and median of blocks

````python
import numpy as np
import matplotlib.pyplot as plt

x=[3,5,7,0,1,4,2] #Original Signal
n=np.arange(-2,5,1) 
plt.figure()
plt.stem(n,x)
plt.title('Original Signal x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')

dsFactor=2
numBlocks = len(x)//dsFactor
x2=np.zeros(numBlocks)
n2=np.zeros(numBlocks)

for i in range(numBlocks):
    x2[i]=np.mean(x[i*dsFactor: (i+1)*dsFactor])
    n2[i]=np.mean(n[i*dsFactor: (i+1)*dsFactor])

plt.figure()
plt.stem(n2,x2)
plt.title('Downsampled Signal x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.show()
````

## Task 3
Upsampling

````python
import numpy as np
import matplotlib.pyplot as plt

x=[3,5,7,0,1,4,2] #Original Signal  
n=np.arange(-2,5,1) 

plt.figure()
plt.stem(n,x)
plt.title('Original Signal x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')

#Upsampling by inserting zero between each samples
usFactor = 2
usSampleSize=len(x)*usFactor

x2=np.zeros(usSampleSize)
x2[::usFactor]=x
n2=np.linspace(n[0], n[-1], usSampleSize)
plt.figure()
plt.stem(n2,x2)
plt.title('Upsample Signal x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')
````

Instead of inserting zeros like this
```
3,5,7,0,1,4,2
3,0,5,0,7,0,0,0,1,0,4,0,2,0
```
we can insert average values or interpolated values
## Task 4 
Upsample by inserting average values

````python
import numpy as np
import matplotlib.pyplot as plt

x=[3,5,7,0,1,4,2] #Original Signal
n=np.arange(-2,5,1) 
plt.figure()
plt.stem(n,x)
plt.title('Original Signal x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')

#Upsampling by inserting average between each samples
usFactor = 2
usSampleSize=len(x)*usFactor
x2=np.zeros(usSampleSize)
n2=np.linspace(n[0], n[-1], usSampleSize)

for i in range(len(n)-1):
    x2[2*i] = x[i]
    x2[2*i+1] = (x[i]+x[i+1])/2

plt.figure()
plt.stem(n2,x2)
plt.title('Upsample Signal x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')
````
