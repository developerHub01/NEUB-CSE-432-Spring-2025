## Task 2 
Downsampling and upsampling image

````python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread("../tnj.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure()
plt.imshow(image)
plt.axis("off")
plt.title('Original image')
plt.show()

#Downsample the image by a factor of 2
downsampledImage = image[::2, ::2]
plt.figure()
plt.imshow(downsampledImage)
plt.axis("off")
plt.title('downsampledImage image')
plt.show()


#upsample the image by a factor of 2
upsampledImage=np.zeros((downsampledImage.shape[0]*2, downsampledImage.shape[1]*2, 3), dtype=np.uint8)
upsampledImage[::2,::2,:]=downsampledImage
plt.figure()
plt.imshow(upsampledImage)
plt.axis("off")
plt.title('upsampledImage image')
plt.show()
````

## Task 3
Downsampling Sound files

This needs soundfile library. If soundfile is not installed, open the anaconda command window and input the following command 
```
pip install soundfile
```

````python
import soundfile as sf
from IPython.display import Audio

y,fs = sf.read("../hello.mp3")
sf.write("hell-v-2.wav", y[::2], fs//2)
sf.write("hell-v-4.wav", y[::4], fs//4)
sf.write("hell-v-8.wav", y[::8], fs//8)
sf.write("hell-v-16.wav", y[::16], fs//16)
sf.write("hell-v-32.wav", y[::32], fs//32)

display(Audio("hell-v-2.wav"))
display(Audio("hell-v-4.wav"))
display(Audio("hell-v-8.wav"))
display(Audio("hell-v-16.wav"))
display(Audio("hell-v-32.wav"))
````

````python
import soundfile as sf
from IPython.display import Audio

y,fs = sf.read("../hello.mp3")
sf.write("hello-reverse.wav", y[::-1], fs)

display(Audio("hello-reverse.wav"))
````
