## Task 1
Generating Unit Impulse Signal

```python
import numpy as np
import matplotlib.pyplot as mpl

n = np.arange(-5, 6, 1)
delta = np.zeros(len(n))
delta[len(n)//2] = 1
mpl.stem(n, delta, "red")
mpl.xlabel("Sample (n)")
mpl.ylabel("$\\delta[n]$")
mpl.title("Unit impulse ($\\delta$) signal")
mpl.show()
```

### Showing assymmetrical delta signal

```python
import numpy as np
import matplotlib.pyplot as mpl

n = np.arange(-5, 11, 1)
delta = np.zeros(len(n))
for i in range(len(n)):
    if(not n[i]): delta[i] = 1

mpl.stem(n, delta)
mpl.title("Unit impulse ($\\delta$) signal")
mpl.xlabel("Sample (n)")
mpl.ylabel("$\\delta$")
mpl.show()
```

## Task 2
Generating Unit Step Signal

```python
import numpy as np
import matplotlib.pyplot as mpl

n = np.arange(-15, 20, 1)
u = np.zeros(len(n))

for index, value in enumerate(n):
    if(value >= 0): u[index] = 1

mpl.stem(n, u)
mpl.title("Unit unit signal")
mpl.xlabel("Sample (n)")
mpl.ylabel("$u[n]$")
mpl.show()
```

## Task 3
Generating Unit Ramp Signal

```python
import numpy as np
import matplotlib.pyplot as mpl

n = np.arange(-5, 10, 1)
u_r = np.zeros(len(n))

for index, value in enumerate(n):
    if value >= 0: u_r[index] = value

mpl.stem(n, u_r)
mpl.title("Unit unit ramp signal")
mpl.xlabel("Sample (n)")
mpl.ylabel("$u_r[n]$")
mpl.show()
```

## Task 4
Generating Exponential Signal

```python
import numpy as np
import matplotlib.pyplot as mpl

n = np.arange(0,11,1)
x = np.zeros(len(n))
a = 1.5

for i in range(len(n)):
    x[i] = a**i

mpl.stem(n, x)
mpl.title("Unit exponential signal")
mpl.xlabel("Sample (n)")
mpl.ylabel("$x[n]$")
mpl.show()  
```
