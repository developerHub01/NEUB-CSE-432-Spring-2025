## Task 1
Manually calculating DFT of a discrete time sequence 

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4])
N = len(x)
X = np.zeros(N, dtype=complex)

for i in range(N):
    for k in range(N):
        X[i] += x[k]*np.exp(-1j*2*np.pi*i*k/N)

print(X)   
```

## Task 2
Calculating DFT of a sinusoid using FFT function of numpy

```python
import numpy as np
import matplotlib.pyplot as plt

f = 50
Fs = 1000
T = 1

t = np.linspace(0, T, int(Fs*T), endpoint=False)
x = np.sin(2*np.pi*f*t)

plt.plot(t[:250], x[:250])
plt.title("Original sinusoid of 50Hz")
plt.xlabel("time t")
plt.ylabel("$sin(2*\\pi*f*t)$")
plt.show()

N = len(x)
X = np.fft.fft(x)
X = np.abs(X[:N//2])/N
frequencies = np.fft.fftfreq(N, d=1/Fs)
frequencies = frequencies[:N//2]
plt.plot(frequencies, X)
plt.title("FFt of sinusoid of 50Hz")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```

## Task 3
Manually calculating DFT of a sinusoid

```python
import numpy as np
import matplotlib.pyplot as plt

f1 = 100
f2 = 200
f3 = 350
Fs = 1000
T = 2
L = 64

t = np.linspace(0, T, int(Fs*T), endpoint=False)
x1 = np.sin(2*np.pi*f1*t)
x2 = np.sin(2*np.pi*f2*t)
x3 = np.sin(2*np.pi*f3*t)
x = x1+x2+x3

plt.figure()
plt.plot(t[:L], x[:L])
plt.title("Original Signal")
plt.xlabel("time t")
plt.ylabel("x(t)")

N = len(x)

Xw = np.zeros(N, dtype=complex)
for i in range(N):
    for k in range(N):
        Xw[i]+=x[k]*np.exp(-1j*2*np.pi*i*k/N)

Y = np.abs(Xw[:N//2])/N
frequencies = np.linspace(0, Fs/2, N//2)

plt.figure()
plt.plot(frequencies, Y)
plt.title("Frequency Spectrum (Manual DFT)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```
