# Task 2
Generating Continuous Time Signal

```python
import numpy as np
import matplotlib.pyplot as mpl

x = np.arange(0, 6*np.pi, 0.05)
n = np.sin(x)

mpl.plot(x, n)
mpl.show()
```

# Task 3
Decorating Plots

```python
import numpy as np
import matplotlib.pyplot as mpl

x = np.arange(0, 6*np.pi, 0.05)
n = np.sin(x)
m = np.cos(x)

mpl.plot(x, n, color="red")
mpl.plot(x, m, color="green")
mpl.title("Sin and cosine curves")
mpl.xlabel("n")
mpl.ylabel("Value")
mpl.show()
```

# Task 4
Plotting multiple signals using subplots.

```python
import numpy as np
import matplotlib.pyplot as mpl

n = np.arange(0, 6*np.pi, 0.01)
x = np.sin(n)
y = np.cos(n)

fig, ax = mpl.subplots(1, 2)

mpl.subplot(1, 2, 1)
mpl.plot(n, x, "red")
mpl.title("Sine Curve")
mpl.subplot(1, 2, 2)

mpl.plot(n, x, "blue")
mpl.title("Cosine Curve")
mpl.suptitle("Sin vs Cosine")
mpl.show()
fig.tight_layout()
```
