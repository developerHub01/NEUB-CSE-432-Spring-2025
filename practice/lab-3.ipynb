## Task 1
Generating Sinusoidal signal

x[n] = A * sin(2πnF/Fs + θ)

```python
import numpy as np
import matplotlib.pyplot as mpl

F = 50 # Analog freequency
A = 10 # Amplitude
theta = 0 # Phase
Fs = 1000 # Sampling freequency
samplingSize = 100

x = np.zeros(samplingSize)

for i in range(len(x)):
    # x[n] = A * sin(2πnF/Fs + θ)
    x[i] = A * np.sin(2*np.pi*i*F/Fs + theta)

mpl.stem(x)
mpl.xlabel('sample(n)')
mpl.ylabel('x[n]')
mpl.title('Sinusoidal signal')
mpl.show()
```

## Task 2
Observing the effect of sampling

```python
import numpy as np
import matplotlib.pyplot as mpl
F=50 #Analog frequency of sinusoid
A=10 #Amplitude of sinusoid
theta=0 #initial phase of sinusoid
sampleSize = 100

fig = mpl.figure(figsize=(10, 15))

Fs = 1000
x = np.zeros(samplingSize)
for n in range(samplingSize):
    x[n] = 2 * A * np.sin(2 * np.pi * n * F/Fs + theta)

mpl.subplot(3, 1, 1)
mpl.stem(x)    
mpl.xlabel('sample(n)')
mpl.ylabel('x[n]')
mpl.title('Sampling Frequency = 1000 hz')

Fs = 500
x = np.zeros(samplingSize)
for n in range(samplingSize):
    x[n] = 2 * A * np.sin(2 * np.pi * n * F/Fs + theta)

mpl.subplot(3, 1, 2)
mpl.stem(x)    
mpl.xlabel('sample(n)')
mpl.ylabel('x[n]')
mpl.title('Sampling Frequency = 500 hz')

Fs = 100
x = np.zeros(samplingSize)
for n in range(samplingSize):
    x[n] = 2 * A * np.sin(2 * np.pi * n * F/Fs + theta)

mpl.subplot(3, 1, 3)
mpl.stem(x)    
mpl.xlabel('sample(n)')
mpl.ylabel('x[n]')
mpl.title('Sampling Frequency = 100 hz')

mpl.show()
fig.tight_layout()
```

## Task 3
Signal time shifting

```python
import numpy as np
import matplotlib.pyplot as plt

#Generating a DT signal
x=[1,2,5,6,9,-5,5,3,0,-1,0,-3,7,6]
n=np.arange(-4,10,1)
plt.figure()
plt.stem(n,x)

m = 4

n = np.arange(-4+m, 10+m, 1)
plt.figure()
plt.stem(n,x)

n = np.arange(-4-m, 10-m, 1)
plt.figure()
plt.stem(n,x)
```

## Task 4
Folding Operation on DT signal

```python
import numpy as np
import matplotlib.pyplot as mpl

#Generating a DT signal
x=[1,2,5,6,9,-5,5,3,0,-1,0,-3,7,6]
n = np.arange(-4, 10, 1)

mpl.figure()
mpl.stem(n, x)

print(n)
n1 = -np.flip(n)
x1 = np.flip(x)
mpl.figure()
mpl.stem(n1, x1)

fig = mpl.figure(figsize=(12, 5))

mpl.subplot(1, 2, 1)
mpl.stem(n, x)
mpl.title("Original signal")
mpl.subplot(1, 2, 2)
mpl.stem(n1, x1)
mpl.title("Folded signal")
mpl.suptitle("Folding a signal")

fig.tight_layout()
```

```python
import numpy as np
x=[1,2,3,1]
h=[1,2,1,-1]

conv = np.zeros(len(x) + len(h) - 1).astype(int)

for i in range(len(x)):
    for j in range(len(h)):
        conv[i+j] += x[i] * h[j]

conv
```
