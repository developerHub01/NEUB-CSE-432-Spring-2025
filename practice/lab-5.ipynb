## Task 1
Frequency analysis of Discrete-Time Periodic Signals (Fourier Series)

````python
import numpy as np
import matplotlib.pyplot as plt

f1=50
f2=100
f3=150
Fs=1000 # Sampling Frequency > (2 * max f)
L=256 # length of the signal
N=512 # Fourier analysis sample points 2L

x = np.zeros(N)

for i in range(L):
    x[i] = np.sin(np.pi*f1*i/Fs) + np.sin(np.pi*f2*i/Fs) + np.sin(np.pi*f3*i/Fs)


plt.figure()
plt.plot(x[0:L])

# Analysis
Ck = np.zeros(N)
for i in range(N):
    for k in range(N):
        Ck[i] += x[k]*np.exp(-1j*2*np.pi*i*k/N)

Y = np.abs(Ck[0:N//2])
plt.figure()
plt.plot(Y)

# Syntesis
x_r = np.zeros(N)
for i in range(N):
    for k in range(N):
        x_r[i] += Ck[k] * np.exp(1j*2*np.pi*i*k/N)

plt.figure()
plt.plot(x_r[0:L])
````

## Task 2
Implementing Zero Order Hold DAC

````python
import numpy as np
import matplotlib.pyplot as plt

# Original digital signal
x = [1, 2, 3, 2, 1]
hold_time = 3  # prottek value koto sample dhore rakhbo

N = len(x) * hold_time
zoh = np.zeros(N)

for i in range(len(x)):
    for j in range(hold_time):
        zoh[i*hold_time + j] = x[i]

plt.plot(zoh, "blue")
plt.stem(x, "red")
````

## Task 3
Cross Correlation of two signals
```
x = [1, 2, 3, 4]
y = [5, 6, 7, 8]
```

````python
import numpy as np
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [5, 6, 7, 8]

cross_corr=np.correlate(x, y, "full")
print(cross_corr)
````
