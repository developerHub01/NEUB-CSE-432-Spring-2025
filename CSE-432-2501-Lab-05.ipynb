## Task 1
Frequency analysis of Discrete-Time Periodic Signals (Fourier Series)



````python
import numpy as np
import matplotlib.pyplot as plt
# Generating Signal
f1=50
f2=100
f3=150
Fs=1000
L=256
N=512

x=np.zeros(N)

for i in range(L):
    x[i]=np.sin(np.pi*f1*i/Fs)+np.sin(np.pi*f2*i/Fs)+np.sin(np.pi*f3*i/Fs)
plt.plot(x[0:L])

#Analysis
Ck=np.zeros(N)
for i in range (N):
    for k in range(N):
        Ck[i]=Ck[i]+x[k]*np.exp(-1j*2*np.pi*i*k/N)
plt.figure()
Y=np.abs(Ck[0:N//2])
plt.plot(Y)

# Synthesis
x_r=np.zeros(N)
for i in range(N):
    for k in range(N):
        x_r[i]=x_r[i]+Ck[k]*np.exp(1j*2*np.pi*i*k/N)
plt.figure()
plt.plot(x_r[0:L])
````
