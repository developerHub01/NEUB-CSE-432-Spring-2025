## Task 1
Generating Unit Impulse Signal

```python
#Showing symmetrical delta signal
import numpy as np
import matplotlib.pyplot as mpl

n = np.arange(-5,6,1)
delta = np.zeros(len(n))
delta[len(n)//2]=1
mpl.stem(n,delta)
mpl.xlabel('Sample (n)')
mpl.ylabel('$\delta[n]$')
mpl.title('Unit Impulse ($\delta$)signal')
mpl.show()
```

```python
#Showing assymmetrical delta signal
import numpy as np
import matplotlib.pyplot as mpl

n = np.arange(-5,11,1)
delta = np.zeros(len(n))
for i in range(len(n)):
    if n[i]==0:
        delta[i]=1

mpl.stem(n,delta)
mpl.xlabel('Sample (n)')
mpl.ylabel('$\delta[n]$')
mpl.title('Unit Impulse ($\delta$)signal')
mpl.show()
```

## Task 2
Generating Unit Step Signal

```python
import numpy as np
import matplotlib.pyplot as mpl

n = np.arange(-5,5,1)
u = np.zeros(len(n))
for i in range(len(n)):
    if n[i]>=0:
        u[i]=1

mpl.stem(n,u)
mpl.xlabel('Sample (n)')
mpl.ylabel('$u[n]$')
mpl.title('Unit step ($u[n]$)signal')
mpl.show()
```

## Task 3
Generating Unit Ramp Signal

```python
import numpy as np
import matplotlib.pyplot as mpl

n = np.arange(-5,11,1)
u_r = np.zeros(len(n))
for i in range(len(n)):
    if n[i]>=0:
        u_r[i]=n[i]

mpl.stem(n,u_r)
mpl.xlabel('Sample (n)')
mpl.ylabel('$u_r[n]$')
mpl.title('Unit ramp ($u_r[n]$)signal')
mpl.show()
```

## Task 4
Generating Exponential Signal

```python
import numpy as np
import matplotlib.pyplot as mpl

n = np.arange(0,11,1)
x = np.zeros(len(n))
a=1.5
for i in range(len(n)):
        x[i]=a**i

mpl.stem(n,x)
mpl.xlabel('Sample (n)')
mpl.ylabel('$x[n]$')
mpl.title('Exponential Signal (a=1.5)')
mpl.show()
```
