# Task 2

# Task 3
Decorating Plots

```python
import numpy as np
import matplotlib.pyplot as mpl
n=np.arange(0,6*np.pi,0.01)
x=np.sin(n)
y=np.cos(n)
mpl.plot(n,x, color='red')
mpl.plot(n,y, color='yellow')
mpl.title("Sine and Cosine curves")
mpl.xlabel("n")
mpl.ylabel("Values")
mpl.show
```

# Task 4
Plotting multiple signals using subplots.

```python
import numpy as np
import matplotlib.pyplot as mpl
n=np.arange(0,6*np.pi,0.01)
x=np.sin(n)
y=np.cos(n)
fig, ax = mpl.subplots(1,2)
mpl.subplot(1,2,1)
mpl.plot(n,x, color='red')
mpl.title("Sine Curve")
mpl.subplot(1,2,2)
mpl.plot(n,y, color='yellow')
mpl.title("Cosine Curve")
mpl.suptitle("Sine and Cosine Curves")
mpl.show
fig.tight_layout()
```
