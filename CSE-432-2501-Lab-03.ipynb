## Task 1
Generating Sinusoidal signal

```python
import numpy as np
import matplotlib.pyplot as mpl
F=50 #Analog frequency of sinusoid
A=10 #Amplitude of sinusoid
theta=0 #initial phase of sinusoid
Fs = 1000 #Sampling Rate
sampleSize=100
x=np.zeros(sampleSize)
for n in range(sampleSize):
    x[n]=A*np.sin(2*np.pi*F/Fs*n+theta)
mpl.stem(x)    
mpl.xlabel('sample(n)')
mpl.ylabel('x[n]')
mpl.title('Sinusoidal signal')
mpl.show
```

## Task 2
Observing the effect of sampling

```python
import numpy as np
import matplotlib.pyplot as mpl
F=50 #Analog frequency of sinusoid
A=10 #Amplitude of sinusoid
theta=0 #initial phase of sinusoid

fig, ax = mpl.subplots(3,1)

mpl.subplot(3,1,1)
Fs = 1000 #Sampling Rate
sampleSize=100
x=np.zeros(sampleSize)
for n in range(sampleSize):
    x[n]=A*np.sin(2*np.pi*F/Fs*n+theta)
mpl.stem(x)    
mpl.xlabel('sample(n)')
mpl.ylabel('x[n]')
mpl.title('Sampling Frequency = 1000 hz')


mpl.subplot(3,1,2)
Fs = 500 #Sampling Rate
sampleSize=100
x=np.zeros(sampleSize)
for n in range(sampleSize):
    x[n]=A*np.sin(2*np.pi*F/Fs*n+theta)
mpl.stem(x)    
mpl.xlabel('sample(n)')
mpl.ylabel('x[n]')
mpl.title('Sampling Frequency = 500 hz')


mpl.subplot(3,1,3)
Fs = 100 #Sampling Rate doesnot follow Nyquist criterion
sampleSize=100
x=np.zeros(sampleSize)
for n in range(sampleSize):
    x[n]=A*np.sin(2*np.pi*F/Fs*n+theta)
mpl.stem(x)    
mpl.xlabel('sample(n)')
mpl.ylabel('x[n]')
mpl.title('Sampling Frequency = 100 hz')

fig.tight_layout()
mpl.show
```

## Task 3
Signal time shifting

```python
import numpy as np
import matplotlib.pyplot as plt

#Generating a DT signal
x=[1,2,5,6,9,-5,5,3,0,-1,0,-3,7,6]
n=np.arange(-4,10,1)
plt.figure()
plt.stem(n,x)

#Delay a signal by m points
m=2
n1=np.arange(-4+m,10+m,1)
plt.figure()
plt.stem(n1,x)


#Advance a signal by m points
m=2
n2=np.arange(-4-m,10-m,1)
plt.figure()
plt.stem(n2,x)

plt.figure()
fig, ax = plt.subplots(3,1)
plt.subplot(3,1,1)
plt.stem(n,x)
plt.title('Original Signal')

plt.subplot(3,1,2)
plt.stem(n1,x)
plt.title('Signal Delayed by 2 points')

plt.subplot(3,1,3)
plt.stem(n2,x)
plt.title('Signal Advanced by 2 points')
fig.tight_layout()
plt.show()
```

## Task 4
Folding Operation on DT signal

```python
import numpy as np
import matplotlib.pyplot as plt

#Generating a DT signal
x=[1,2,5,6,9,-5,5,3,0,-1,0,-3,7,6]
# n=np.arange(-4,10,1)
n=np.arange(len(x))-4
plt.figure()
plt.stem(n,x)

#Folding
x1=np.flip(x)
n1=np.arange(len(x))-9
plt.figure()
plt.stem(n1,x1)

plt.figure(figsize=(1024,720))
fig, ax=plt.subplots(1,2)
plt.subplot(1,2,1)
plt.stem(n,x)
plt.title('Original Signal')
plt.subplot(1,2,2)
plt.stem(n1,x1)
plt.title('Folded Signal')
fig.tight_layout()
fig.savefig("figure.pdf")
plt.show()
```

## Task 5
Convolution

```python
#Convolution using NUmpy built-in function 
import numpy as np
x=[1,2,3,1]
h=[1,2,1,-1]
y=np.convolve(x,h)
print(y)
z=np.convolve(h,x)
print(z)
```

```python
#Manually coding convolution
import numpy as np
x=[1,2,3,1]
h=[1,2,1,-1]
m=len(x)+len(h)-1
y=np.zeros(m).astype(int)
for i in range (len(h)):
    for j in range (len(x)):
        y[i+j]=y[i+j]+x[j]*h[i]
print(y)        
```
