## Task 1
Downsampling

````python
import numpy as np
import matplotlib.pyplot as plt

x=[3,5,7,0,1,4,2] #Original Signal
n=np.arange(-2,5,1) 
plt.figure()
plt.stem(n,x)
plt.title('Original Signal x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')

#Down-Sampling
dsFactor=2
y=x[::dsFactor]
k=n[::dsFactor]
plt.figure()
plt.stem(k,y)
plt.title('DownSampled Signal y[n]')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.show()
````

This downsampling method just omits every other signal. Another method of downsampling can be done by grouping the signals into blocks and finding either mean or median.
```
Original Signal              3,5,7,0,1,4,2
Signals Divided into blocks  3,5   7,0  1,4  
Finding average of each block 4    3.5  2.5  
```
### Task 2 
Do downsampling by using mean and median of blocks

````python
import numpy as np
import matplotlib.pyplot as plt

# Original signal
x = [3, 5, 7, 0, 1, 4, 2]
n = np.arange(-2, 5, 1)  # origin = 2
plt.figure()
plt.stem(n, x)
plt.title('Original Signal x(n)')
plt.xlabel('n')
plt.ylabel('x(n)')

# Downsampling by averaging over blocks
block_size = 2  # Size of each block
num_blocks = len(x) // block_size  # Number of complete blocks
x_downsampled = [np.mean(x[i * block_size:(i + 1) * block_size]) for i in range(num_blocks)]
n_downsampled = np.arange(num_blocks)  # Downsampled indices

plt.figure()
plt.stem(n_downsampled, x_downsampled)
plt.title('Downsampled Signal (Averaged Over Blocks)')
plt.xlabel('Block Index')
plt.ylabel('Average Value')

plt.show()
````

## Task 3
Upsampling

````python
import numpy as np
import matplotlib.pyplot as plt

x=[3,5,7,0,1,4,2] #Original Signal
n=np.arange(-2,5,1) 
plt.figure()
plt.stem(n,x)
plt.title('Original Signal x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')

#Upsampling by inserting zero between each samples
usFactor = 2
usSignalLength= len(x)*usFactor
y=np.zeros(usSignalLength)
y[::usFactor]=x
# k=np.arange(-2,5,1/usFactor)
k=np.linspace(n[0],n[-1],usSignalLength)
plt.figure()
plt.stem(k,y)
plt.title('UpSampled Signal y[n]')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.show()
````

Instead of inserting zeros like this
```
3,5,7,0,1,4,2
3,0,5,0,7,0,0,0,1,0,4,0,2,0
```
we can insert average values or interpolated values
## Task 4 
Upsample by inserting average values

````python
import numpy as np
import matplotlib.pyplot as plt

x=[3,5,7,0,1,4,2] #Original Signal
n=np.arange(-2,5,1) 
plt.figure()
plt.stem(n,x)
plt.title('Original Signal x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')

#Upsampling by inserting average between each samples
usFactor = 2
usSignalLength= len(x)*usFactor
y=np.zeros(usSignalLength)
k=np.linspace(n[0],n[-1],usSignalLength)

for i in range (len(x)-1):
    y[2*i]=x[i]
    y[2*i+1]=(x[i]+x[i+1])/2
y[-1]=x[-1]    

plt.figure()
plt.stem(k,y)
plt.title('UpSampled Signal y[n]')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.show()
````
